## Factorial calculations

```python
def fact(n):
    '''
    This function returns a factorial of n. 4! = 1*2*3*4; n! = 1*2*...*n
    input: an integer
    output: n!
    '''
    if int(n)!=n:
        return 'This is not an integer!'
    r = 1
    while n>0:
        r = r*n
        n = n -1 
    return r
```

```python
fact(100)
```

## Concatenating text

```python
def add_words(*words):
    res = ''
    for word in words:
        res  = res + word+' '
    res = res.strip()
    return res

add_words('Social', 'distancing', 'saves', 'lives')
```

## Exception handling

```python
def ratio(x,y):
    '''
    This function takes two numbers (x and y) and returns x/y.
    input: x, and y
    output: x/y
    '''
    try:
        return x/y
    except Exception as err:
        print(err)
```

```python

def divide(x,y):
    try:
        return x/y
    except ZeroDivisionError:
        print('Are you dividing by zero???')
    except TypeError as err:
        print(err)
    except Exception as err:
        print('Something went bad')
    else:
        print('result')
    finally:
        print('Excution Completed')
        
divide(5,4)
divide(5,0)
divide(5,'a')
```

```python
import string
def nchars(text):
    '''
    Function that calculates the total number of characters EXCLUDING punctuations
    '''
    n = 0
    for ch in text:
        if (ch not in string.punctuation) and (ch!=' ') and (ch not in string.digits):
            n = n + 1
    return n
```

```python
assert nchars('abd *%*&^&^*&^*&^*&') ==3, 'Incorrect output'
assert nchars('$$$abd&&&nn&&$#@')==5, 'Incorrect output 2'
assert nchars('*&^*&^ABC  876876 987987&&^*&^')==3, 'Incorrect output 3'
```
